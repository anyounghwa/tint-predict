## [단계 ⑤] 모델의 입력 데이터 형태 설정

```python
# 당일 발주 수량과 가장 유의미한 관계가 있는 변수로 D+3, D+4, D+5일의 투입 예정 수량을 최종 선택하였고, 
# 이를 기반으로 3~5일 전 발주 및 계획 발주 수량을 활용하여 재고관리 최적화를 위한 발주량 예측 모델을 설계한다. 
# 즉, 해당 변수를 활용하여 학습된 예측 모델은 부품의 발주 수량을 3일 전에 예측할 수 있게 된다.

data_94 = df_94.loc[:, ['D일 투입예정 수량(D일계획)','D+3일 투입예정 수량(Total)', 
                        'D+4일 투입예정 수량(Total)','D+5일 투입예정 수량']].reset_index(drop=True)
data_94
```

```python
import numpy as np

# 사용자 정의 함수: Tabular 데이터를 시계열 데이터로 변환
def to_timeseries_data(data_94, lookback=3, delay=3):
    
    # 샘플 수 계산
    output_len = len(data_94) - (lookback + delay) + 1   #49-(3+3)+1 = 44, 44개의 샘플을 만들 수 있음
    
    # feature 개수 (열 개수)
    n_feature = data_94.shape[-1]  #4, 음수 인덱스를 사용하면 뒤에서부터 접근, 마지막차원의 크기

    # 입력과 타겟 초기화
    inputs = np.zeros((output_len, lookback, n_feature))  #(샘플 수, lookback, feature 수), 배열을 만든다
    targets = np.zeros((output_len,))  # (샘플 수,)

    # 데이터 채우기
    for i in range(output_len):
        # lookback만큼 과거 데이터
        inputs[i] = data_94.iloc[i:i+lookback, :].values  #데이터 프레임을 array배열로 만든다.
        # delay 이후 target 값 (첫 번째 열 값 사용)
        targets[i] = data_94.iloc[i+lookback+delay-1, 0]

    return inputs, targets


# data_94 : DataFrame (원본 데이터)
# lookback=3 : 입력으로 사용할 과거 일 수 (default=3)
# delay : target으로 사용할 미래 일 수 (default=3)


# inputs[i] = data_94.iloc[i:i+lookback, :]  과거 3일치
# inputs[i] =  i번째부터 i+lookback까지 슬라이싱 → 3일치 데이터 채워 넣기
# targets[i]:i+lookback+delay-1 번째 행의 첫 번째 열 값 → 미래 예측 값으로 사용
```

```python
data_94.shape[-1]
```

```python
# 예시로 df_94에서 일부 열 선택
data_94 = df_94.loc[:, ['D일 투입예정 수량(D일계획)','D+3일 투입예정 수량(Total)', 
                        'D+4일 투입예정 수량(Total)', 'D+5일 투입예정 수량']].reset_index(drop=True)
# 함수 적용
X_94, y_94 = to_timeseries_data(data_94, lookback=3, delay=3)

# 결과 확인
print("X의 shape:", X_94.shape)  # (샘플 수, 타임스텝, 피쳐쳐 수)
print("y의 shape:", y_94.shape)  # (샘플 수,)
```

```python
# parameters = (input_dim + units + 1) × units × 4
# 4: LSTM은 4개의 게이트 (input, forget, cell, output 게이트)를 가짐
# units: 출력 유닛 수
# input_dim: 입력 피처 수


# 첫 번째 LSTM 레이어 = LSTM(8, dropout=0.2, activation='relu', input_shape=(3,4), return_sequences=True)
# units = 8, input_dim = 4
# (4+8+1)×8×4  = 416

# 두 번째 LSTM 레이어 = (8+8+1)*8*4 = 544

# Dense 레이어
# parameters=(input_dim+1)×output_dim
# Param = (8+1) * 1 = 9
```

```python
(8+8+1)*8*4
```

```python
#학습방식 설정, 
#모델은 Adam으로 파라미터를 업데이트하고 MAE 기준으로 얼마나 잘 맞췄는지 판단하며 학습한다
model.compile(optimizer='adam', loss='mae')

#loss='mae' : MAE (Mean Absolute Error) = 평균 절대 오차
```

```python
# import os
# os.environ['TMPDIR'] = os.path.abspath('./temp')

# # temp 폴더 생성
# os.makedirs('./temp', exist_ok=True)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import os

os.makedirs('./model', exist_ok=True)

model_path = './model/{epoch:02d}-{val_loss:.4f}.keras'
#{epoch:02d} : 현재 epoch 숫자를 두 자리 정수(예: 01, 02, 03...)로 저장
#{val_loss:.4f} : validation loss 값을 소수점 4자리까지 저장
#만약 7번째 epoch에서 validation loss가 0.123456일때 => ./model/07-0.1235.keras로 저장된다.

callbacks = [EarlyStopping(monitor='val_loss', patience=15),
             ModelCheckpoint(filepath=model_path, 
                             monitor='val_loss', 
                             verbose=0, 
                             save_best_only=True,
                             save_weights_only=False)]

history = model.fit(X_train_94, #입력 데이터(특징, 피처들)
                    y_train_94,  #타킷 데이터(정답, 예측할 값)
                    epochs=100,  #최대 학습 반복 횟수(에포크)
                    batch_size=4,  #한 번에 학습할 데이터 수 
                    validation_data=(X_val_94, y_val_94), #검증 데이터
                    callbacks=callbacks)  #조기종료, 체크포인트 저장 등


# save_weights_only=False 모델 전체를 저장하겠다.
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epoch = history.epoch

plt.figure()
plt.plot(epoch, loss, 'r', label='Training loss')
plt.plot(epoch, val_loss, 'g', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.show()
```
